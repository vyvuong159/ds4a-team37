### Recover missing zip codes from the same hospital

# 2. Deal with incorrect data type

'Length of Stay' should be numeric, but it's currently an object, let's have a look at the variable

```python
df_clean['Length of Stay'].unique()
```

Because there is "120+" value --> object. Let treat "120+" as 120 and convert it to numeric

```python
df_clean['Length of Stay'] = df_clean['Length of Stay'].apply(lambda x: str(x).split(' ')[0])
df_clean['Length of Stay'] = pd.to_numeric(df_clean['Length of Stay'])
df_clean['Length of Stay'].unique()
```

Birth Weight should be numeric, but it's an object. Let's look at the variable

```python
df_clean['Birth Weight'].unique()
```

Because it contains "UNKN" and also weird presentation of number, like 04000. We can:
- Convert UNKN (Unknown) to NULL
- Convert all other to numeric


```python
df_clean_1 = df_clean
df_clean_1['Birth Weight_clean'] = df_clean_1['Birth Weight'].apply(pd.to_numeric, errors='coerce')
df_clean_1['Birth Weight_clean'].unique()
```

```python
df_clean_1[df_clean_1['Birth Weight'] == 'UNKN']
df_clean_1.info()
```

```python
#Rename Birth Weight_clean column to "Birth Weight" and drop the original column
df_clean_2 = df_clean_1.drop(['Birth Weight'], axis=1).rename(columns={"Birth Weight_clean": "Birth Weight"})
df_clean_2.info()
```

Investigate potential categorical variables

```python
df_clean['APR Risk of Mortality'].value_counts()
```

```python
df_clean['Payment Typology 1'].value_counts()
```

Create new columns to transform categorical variables

```python
age_dict = {'0 to 17': 1, '18 to 29': 2, '30 to 49': 3, '50 to 69': 4, '70 or Older': 5}

mortality_dict = {'Minor': 1, 'Moderate': 2, 'Major': 3, 'Extreme': 4}

payment_typo_dict = {'Medicare': 1, 'Medicaid': 2, 'Private Health Insurance': 3, 
                     'Blue Cross/Blue Shield': 4, 'Self-Pay': 5, 'Managed Care, Unspecified': 6, 
                     'Miscellaneous/Other': 7, 'Federal/State/Local/VA': 8, 
                     'Department of Corrections': 9, 'Unknown': 9999}

apr_medical_surgical_code = {'Medical': 1, 'Surgical': 2}

df_clean_2['age_group_code'] = df_clean_2['Age Group'].apply(lambda x: age_dict[x])
df_clean_2['apr_risk_mortality_code'] = df_clean_2['APR Risk of Mortality'].apply(lambda x: mortality_dict[x])
df_clean_2['payment_typology_1_code'] = df_clean_2['Payment Typology 1'].apply(lambda x: payment_typo_dict[x])
df_clean_2['apr_medical_surgical_code'] = df_clean_2['APR Medical Surgical Description'].\
                    apply(lambda x: apr_medical_surgical_code[x])


df_clean_2.head()
```

```python
df_clean_2.isna().sum()
```

Transform from numeric to string 

```python
df_clean_2[["Operating Certificate Number", "Permanent Facility Id"]] = df_clean_2[["Operating Certificate Number", 
                                                                                    "Permanent Facility Id"]].astype(str) 
df_clean_2.info()
```

### Prepare the data for modeling purpose
- Remove rows with NAs 
- Replace OOS character in the zip code (out of state) with numeric variable 9999
- Remove all unnecessary columns
- Transform all should-have-been-categorical variables to category

```python
#Remove rows with NAs 
df_modeling = df_clean_2.dropna(subset=['Hospital Service Area'])
```

```python
#Replace OOS character in the zip code (out of state) with numeric variable 9999
df_modeling['Zip Code - 3 digits_clean'] = df_modeling['Zip Code - 3 digits'].apply(lambda x: str(x).replace('OOS','9999'))  
df_modeling['Zip Code - 3 digits_clean'] = pd.to_numeric(df_modeling['Zip Code - 3 digits_clean'])
```

```python
#Transform all should-have-been-categorical variables to category
df_modeling['CCS Diagnosis Code'] = df_modeling['CCS Diagnosis Code'].astype('category')
df_modeling['CCS Procedure Code'] = df_modeling['CCS Procedure Code'].astype('category')
df_modeling['APR DRG Code'] = df_modeling['APR DRG Code'].astype('category')
df_modeling['APR MDC Code'] = df_modeling['APR MDC Code'].astype('category')
df_modeling['APR Severity of Illness Code'] = df_modeling['APR Severity of Illness Code'].astype('category')
df_modeling['age_group_code'] = df_modeling['age_group_code'].astype('category')
df_modeling['apr_risk_mortality_code'] = df_modeling['apr_risk_mortality_code'].astype('category')
df_modeling['payment_typology_1_code'] = df_modeling['payment_typology_1_code'].astype('category')
df_modeling['apr_medical_surgical_code'] = df_modeling['apr_medical_surgical_code'].astype('category')
```
